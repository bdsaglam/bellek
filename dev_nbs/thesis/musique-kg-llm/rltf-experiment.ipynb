```python
import pandas as pd
```

```python
df = pd.read_csv('./rltf-experiment-logs.csv')
df.head()
```

```python
df = df.rename(columns={"Step": 'step', 'misunderstood-star-13 - env/reward_mean': 'mean_reward'})[['step', 'mean_reward']].copy()
df.head()
```

```python
df.to_json('./rltf-experiment-logs.jsonl', orient='records', lines=True)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Convert pandas columns to numpy arrays
steps = df['step'].values

# Try plotting using numpy arrays
plt.figure(figsize=(8, 6))
plt.plot(steps, df['mean_reward'], label='Mean Reward', color='blue')

plt.xlabel('Step', fontsize=14)
plt.ylabel('Mean Reward', fontsize=14)
# plt.title('Mean Reward for Reinforcement Learning Model', fontsize=16)
# plt.legend(fontsize=12)
plt.grid(True)

plt.savefig('reward_plot_numpy.png', format='png', dpi=300)
plt.show()
```
