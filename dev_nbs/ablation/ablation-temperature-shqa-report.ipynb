```python
pd.options.display.float_format = '{:,.3f}'.format
```

```python
df = pd.read_json("ablation-temperature-shqa-results-20240822-081554.jsonl", lines=True)
df.rename(
    columns={
        'exact_match': 'EM',
        'f1': 'F1',
    },
    inplace=True,
)
column_order = ['Prompting', 'Temperature', 'Run', 'EM', 'F1']
df = df.loc[:, column_order].copy().sort_values(['Prompting', 'Temperature', 'Run']).reset_index(drop=True)
df
```

```python
with open("ablation-temperature-shqa-results-detailed.tex", 'w') as f:
    f.write(df.to_latex(index=False, float_format='%.3f'))
```

```python
agg_df = df.drop(columns=['Run']).groupby(['Prompting', 'Temperature']).mean()
agg_df
```

```python
with open("ablation-temperature-shqa-results-agg.tex", 'w') as f:
    f.write(agg_df.to_latex(float_format='%.3f'))
```
